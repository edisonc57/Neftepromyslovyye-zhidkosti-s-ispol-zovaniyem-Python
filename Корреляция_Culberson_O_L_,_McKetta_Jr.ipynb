```python
import numpy as np
import matplotlib.pyplot as plt

def Rswp_equation(P,T):
    A = 8.15839-6.12265E-02*T+1.91663E-04*T**2.0-2.1654E-07*T**3.0
    B = 1.01021E-02-7.44241E-05*T+3.05553E-07*T**2.0-2.94883E-10*T**3.0
    C = (-9.02505+0.130237*T-8.53425E-04*T**2.0+2.34122E-06*T**3.0-2.37049E-09*T**4.0)*1E-07
    Rswp= A + B*P + C*P**2 
    return (Rswp) 

press= np.linspace(1000, 10000, 10)
temp= np.linspace(80, 340, 20)
Rswp= np.zeros(len(temp))

plt.figure("Rsw",figsize= (4,5))
plt.rcParams.update({'font.size':10})

Rswp	[::] = Rswp_equation(press[0], temp [::]) #En resumen, esta expresión asigna a todos los elementos de la lista Rswp el resultado de la función 
#Rswp_CulbersonMcketta aplicada a cada elemento de la lista Temp y el primer elemento de la lista press.
plt.plot(temp, Rswp, 'k--')

for i in range (0, len(press)):
    Rswp=Rswp_equation(press[i], temp)
    plt.plot(temp,Rswp, 'k--')
    
```

```python
def Rsw_equation (P,T,S):
    A=8.15839-6.12265E-02*T+1.91663E-04*T**2.0-2.1654E-07*T**3.0
    B=1.01021E-02-7.44241E-05*T+3.05553E-07*T**2.0-2.94883E-10*T**3.0
    C=(-9.02505+0.130237*T-8.53425E-04*T**2.0+2.34122E-06*T**3.0-2.37049E-09*T**4.0)*1E-07
    Rswp= A + B*P + C*P**2 
    Rsw = 10**(-0.0840655*S*T**-0.285854)*Rswp
    return (Rsw)

press= np.linspace(1000, 10000,10)
temp= np.linspace(80, 340, 20)
Rsw= np.zeros(len(temp))
Rswp= np.zeros(len(temp))

plt.figure("Rsw", figsize=(4,5))
plt.rcParams.update({'font.size':10})

for i in range (0, len(press)):
    Rswp=Rswp_equation(press[i], temp)
    Rsw=Rsw_equation(press[i], temp, 3.1 )
    plt.plot(temp,Rswp, 'k--')
    plt.plot(temp,Rsw, 'r-')
    plt.title('Корреляция Culberson O.L., McKetta-Jr', fontsize=11, color= [1,0,1])
    plt.xlabel('Temperatura °F', fontsize= 11, color= 'red' )
    plt.ylabel("Rwsp", fontsize=11, color= 'blue')
    plt.grid('True')
    plt.minorticks_on()
    plt.grid(visible=True, which='minor', color= 'g', linestyle= '--')
    plt.savefig('Корреляция Culberson O.L., McKetta-Jr.png', dpi=300)

fig, ax1 = plt.subplots(figsize=(4,5))
color= 'tab:green'
ax1.set_xlabel('Temperatura °F')
ax1.set_ylabel('Rswp', color=color)

for i in range (0, len(press)):
    Rswp[::]=Rswp_equation(press[i],temp[::])
    ax1.plot(temp, Rswp, 'r--')
    ax1.tick_params(axis='y',labelcolor=color)  

ax2=ax1.twinx()
color= 'tab:blue'
ax2.set_ylabel('Rws', color=color)


for i in range (0, len(press)):
    Rsw[::]=Rsw_equation(press[i],temp[::],3.1)
    ax1.plot(temp, Rsw, 'b-')
    ax1.tick_params(axis='y',labelcolor=color)  
```
