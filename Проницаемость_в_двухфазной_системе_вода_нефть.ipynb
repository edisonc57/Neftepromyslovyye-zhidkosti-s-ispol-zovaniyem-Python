```python
# -*- coding: utf-8 -*-
"""
Spyder Editor

This is a temporary script file.
"""

import matplotlib.pyplot as plt
import numpy as np

def krw(Sw, Srw, Sro,omega):
    Se= (Sw-Srw)/(1-Srw-Sro)
    krw= Se**omega
    return krw

def kro(Sw, Srw, Sro,omega):
    Se= (Sw-Srw)/(1-Srw-Sro)
    kro= (1-Se)**omega
    return kro

omega= 2
Sw= 0.5
Srw= 0.15
Sro= 0.1
krw_point= krw(Sw, Srw, Sro, omega)


Sw_vector= np.linspace (0.15, 1-Sro, 100)
krw_vector= krw(Sw_vector, Srw, Sro, omega)
kro_vector= kro(Sw_vector, Srw, Sro, omega)

plt.plot(Sw_vector, kro_vector,'r-',label='$k_{ro}$')
plt.plot(Sw_vector, krw_vector,'b--',label= '$k_{rw}$')
plt.xlabel("$S_{w}$")
plt.ylabel("$k_{r}$")
plt.legend(loc=0)
plt.title("Типичный вид кривых относительной проницаемости в двухфазной системе вода - нефть")
plt.grid()

```
