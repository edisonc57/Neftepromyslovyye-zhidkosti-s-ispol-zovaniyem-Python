```python
# -*- coding: utf-8 -*-
"""
Created on Mon Apr 24 21:17:52 2023

@author: Edison
"""

import matplotlib.pyplot as plt
import numpy as np

def krw(Sw, Srw, Sro,omega):
    Se= (Sw-Srw)/(1-Srw-Sro)
    krw= Se**omega
    return krw

def kro(Sw, Srw, Sro,omega):
    Se= (Sw-Srw)/(1-Srw-Sro)
    kro= (1-Se)**omega
    return kro

def Pcow (Pt,Sw, Srw, Sro, tetha):
    Se= (Sw-Srw)/(1-Srw-Sro)
    Pcow= Pt*Se**(-1/tetha)
    return Pcow
    

omega= 2
Sw= 0.5
Srw= 0.15
Sro= 0.1
thetha= 2
Pt= 1e04
krw_point= krw(Sw, Srw, Sro, omega)

Sw_vector= np.linspace (Srw, 1-Sro, 100)
krw_vector= krw(Sw_vector, Srw, Sro, omega)
kro_vector= kro(Sw_vector, Srw, Sro, omega)

plt.figure(1)
plt.plot(Sw_vector, kro_vector,'r-',label='$k_{ro}$')
plt.plot(Sw_vector, krw_vector,'b--',label= '$k_{rw}$')
plt.xlabel("$S_{w}$")
plt.ylabel("$k_{r}$")
plt.legend(loc=0)
plt.title("         - ")
plt.grid()

plt.figure(2)
Sw_vector= np.linspace (Srw+1e-02, 1-Sro, 100)
pcow_vector= Pcow(Pt, Sw_vector, Srw, Sro,thetha )
plt.figure(2)
plt.plot(Sw_vector, pcow_vector,'m--', label= '$p_{cow}$' )
plt.xlabel('$s_{w}$')
plt.ylabel("$P_{cow}$ {Pa}")
plt.legend(loc=0)
plt.title("  ")
plt.grid()
```
